```python
source = 'kalyan'
dest = 'thane'
```

```python
import requests
from bs4 import BeautifulSoup
import urllib
print ('http://dev.virtualearth.net/REST/V1/Routes/Driving?o=xml&wp.0=%s&wp.1=%s&avoid=minimizeTolls&key=AskxPwNie4zbQuyJWIaabkDFc5qffwhKpgBZP6g8o1z3mmc33h81zxIPe8xaM1vF'%(source,dest))
page = requests.get('http://dev.virtualearth.net/REST/V1/Routes/Driving?o=xml&wp.0=%s&wp.1=%s&avoid=minimizeTolls&key=AskxPwNie4zbQuyJWIaabkDFc5qffwhKpgBZP6g8o1z3mmc33h81zxIPe8xaM1vF'%(source,dest))
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# soup_text = soup.text
soup_text = str(soup)
```

```python
soup_text
```

```python
dist = soup.find_all("traveldistance")
dist
```

```python
import re
m=re.compile('<traveldistance>(.*?)</traveldistance>', re.DOTALL).findall(soup_text)
m
```

```python
fdist = str(m[0])
```

```python
print ("%s to %s Distance is %s kilometers"%(source,dest,fdist))
```
